## 6. Escribe una funci贸n que devuelva el factorial de un n煤mero.

```python
def factorial(n):
#tu codigo aqu铆
    fac = 1  
    
    for i in range(1,n+1):
        
        fac *= (n- (i-1))  
        
    return fac
```

```python
# F贸rmula factorial
# n! = n * (n - 1) *...*1

# Este c贸digo define una funci贸n llamada "factorial" que toma una entrada "n". La funci贸n utiliza un bucle for para iterar a trav茅s del rango de n煤meros
# desde 1 hasta n+1. Para cada n煤mero en ese rango, multiplica el valor actual de x por el n煤mero en el rango. Al final del bucle,
# la funci贸n devuelve el valor final de x, que ser谩 el factorial del n煤mero de entrada "n".

# El factorial de un entero positivo n es el producto de todos los enteros positivos menores o iguales a n.
# Por ejemplo, el factorial de 6 (escrito "6!") es 6 * 5 * 4 * 3 * 2 * 1 = 720.

# As铆 que esta funci贸n toma una entrada de cualquier entero positivo y devuelve el factorial de ese n煤mero.
```

## 10. Escribe una funci贸n para verificar si una cadena es un pangrama, es decir: si contiene todas las letras del alfabeto al menos una vez. Ten en cuenta que las cadenas pueden contener caracteres que no son letras.

```python
import re
import string

def pangram(string):
#tu codigo aqu铆
    string = string.lower()  
    letras = re.findall(r'[a-z]', string)
    letras_unicas = set(letras)
    if len(letras_unicas) == 26:
        return print("Es una Pangrama" )
    else:
        return print("No es una Pangrama")
```

## 11. Escribe una funci贸n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab茅ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! お`

```python
def sort_alpha(string):
#tu codigo aqu铆
    palabras = string.split(',')

    palabras.sort()

    orden = ', '.join(palabras)

    return orden
```

## 12. Escribe una funci贸n para verificar si una contrase帽a dada es fuerte (al menos 8 caracteres, al menos una min煤scula, al menos una may煤scula, al menos un n煤mero y al menos un car谩cter especial). Deber铆a devolver True si es fuerte y False si no lo es.
